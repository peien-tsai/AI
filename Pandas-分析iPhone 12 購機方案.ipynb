```python
import pandas as pd
dfs = pd.read_html('https://www.cht.com.tw/home/apple/iphone/index')
```

```python
df = dfs[0]  #所有的購機方案
```

```python
df.head()
```

```python
df[df['月繳金額.1'].str.contains('個月')]  #找出月繳金額.1這個欄位裡有出現*個月的資料筆數
```

```python
df['month'] = df['月繳金額.1'].str.extract('(\d+)個月') #將數字部分填入一個叫month的欄位裡
```

```python
df['month'] = df['month'].ffill() #該欄位下有很多都是空值，但契約長度與第一筆相同，故向後填滿
```

```python
#DataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)  
#參考:https://www.itread01.com/content/1544495051.html 參數說明
df.dropna(inplace=True) #inplace=True直接在原資料做刪除整列
```

```python
df.head()
```

```python
#有一列全部資料為個月，要將其過濾
df2 = df[~df['月繳金額.1'].str.contains('個月')] #加上波浪，排除月繳金額.1裡，有幾個月資料的筆數
```

```python
df2.head()
```

```python
#處理欄位名稱
df2.columns
```

```python
df2.columns = ['購機優惠價', '機型', '容量', '$599', '$799', '$999', '$1199', '$1,399',
       '$1,599', '$1,799', '$2,699', 'month']
```

```python
df2.head()
```

```python
#把有$的不同欄位變成，一個欄位裡的資訊
#df2裡有包含$的做轉換，i是要留下的欄位，j是新欄位名稱，suffix是指定要填入什麼(數字)
df3 = pd.wide_to_long(df2,['$'],i = ['購機優惠價','機型','容量','month'],j = 'monthly_price',suffix = '[\d+,]+') 
```

```python
#將索引變成欄位
df3.reset_index(inplace=True)
```

```python
df3.head()
```

```python
#將$欄位裡有-的欄位排除
df4 = df3[df3['$'] != '-']
```

```python
df4.head()
```

```python
#查看一下欄位資料的型態，會發現month,monthly_price與$為字串型態
df4.info()
```

```python
#把month跟$轉為整數型態
df4['month'] = df4['month'].astype(int)
df4['$'] = df4['$'].astype(int)
```

```python
#把monthly_price裡，逗號去除，並將資料轉為整數型態
df4['monthly_price'] = df4['monthly_price'].map(lambda e : int(e.replace(',','')))
```

```python
df4.head()
```

```python
#算出手機的成本價格 = 手機價格+合約價格
df4['total_price'] = df4['$'] + df4['month'] * df4['monthly_price']
```

```python
df4.head()
```

```python
#將總成本由小到大排序
df4.sort_values('total_price')
```
