```python
#Matplotlib pyplot
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi,np.pi,200) #等差數列
y = np.sin(x)

#原始的圖案
plt.plot(x,y)
plt.show()

#線條 顏色 寬度 樣式
plt.plot(x,y,color='blue',linewidth=1.0,linestyle='--')

#增加坐標軸長度
plt.xlim(x.min()*1.2,x.max()*1.2)
plt.ylim(y.min()*1.2,y.max()*1.2)

#座標刻度(Latex)  tick(刻度)
plt.xticks((-np.pi,-np.pi/2,np.pi/2,np.pi),(r'$-\pi$',r'$-\pi/2$',r'$\pi/2$',r'$-\pi$'))
plt.yticks([-1,-0.5,0,0.5,1])

g = plt.gca() #取得目前座標軸

g.spines['right'].set_color('none') #隱藏右框線
g.spines['top'].set_color('none') #隱藏上框線

g.xaxis.set_ticks_position('bottom') #設定x軸刻度在下側    axix(坐標軸)
g.spines['bottom'].set_position(('data',0)) #設定x軸位置

g.yaxis.set_ticks_position('left') #設定y軸刻度在左側
g.spines['left'].set_position(('data',0)) #設定y軸位置

#標出特定的點
t = 3*np.pi/4
plt.plot([t,t],[0,np.sin(t)],color='red',linewidth=1.5,linestyle=':')  #linestyle有 ':'   '-.'   '--'    '-'
plt.scatter(t,np.sin(t),20,color='red')

print(np.sin(t))

#設定點的說明(Latex)  annotate(特定點的註解)  data(數據曲線)
plt.annotate(r'$ sin(\frac{3\pi}{4})=\frac{\sqrt{2}}{2} $', xy=(t, np.sin(t)), xycoords='data',
             xytext=(-90, -50), textcoords='offset points',fontsize='10',
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2'))

#設定坐標軸刻度文字
for label in g.get_xticklabels()+g.get_yticklabels():
    label.set_fontsize(8)
    label.set_color('purple')
    label.set_bbox(dict(facecolor='white',edgecolor='None',alpha=.65))
    
plt.show()
```

```python
#Matplotlib lengend 圖例
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2,100)

plt.plot(x,x,label='linear')
plt.plot(x,x**2,label='quadratic') #次方
plt.plot(x,x**3,label='cubic')  #三次方
plt.plot(x,np.exp(x),label='exponential') #指數

plt.xlabel('x label') #x軸標籤
plt.ylabel('y label') #y軸標籤
plt.title("Simple Plot")  #標題

plt.legend(loc='best') #找到最好的位置(下面補充)

plt.show()


# best        最適宜   0
# upper right 右上角   1
# upper left  左上角   2
# lower right 右下角   3
# lower left  左下角   4
# right       右側     5
# center left 左側中間 6
# center right右側中間 7
# lower center下方中間 8
# upper center上方中間 9
# center      最適宜  10

```

```python
#Matplotlib Figure    
# figure(num=None, figsize=None, dpi=None, facecolor=None, edgecolor=None, frameon=True)
# num:圖像編號或名稱，數字為編號，字符串為名稱
# figsize:指定figure的寬和高，單位為英寸；
# dpi參數指定繪圖對象的分辨率，即每英寸多少個像素，缺值為80
# facecolor:背景顏色
# edgecolor:邊框顏色
# frameon:是否顯示邊框
    
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi,np.pi,200)
y1 = np.cos(x)
y2 = np.sin(x)
y3 = np.tanh(x) #雙曲正切函數

plt.figure(num = 'f1',figsize=(16,4))
plt.plot(x,y1)

plt.figure(num = 'f2',figsize=(16,4))
plt.plot(x,y2)

plt.figure(num = 'f1')
plt.plot(x,y3) #將y3加入f1

plt.show()


```

```python
#Matplotlib subplot  創建一個畫像(figure)和一組子圖( subplots)
#subplot(nrows,ncols,sharex,sharey,subplot_kw,**fig_kw)
#nrows:subplot的行數
#ncols:subplot的列數
#sharex(y):可刪除網格上的內標籤
#subplot_kw:字典類型，可選參數。把字典的關鍵字傳遞給add_subplot()來創建每個子圖
#**fig_kw:把所有詳細的關鍵字參數傳給figure()函數


import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi,np.pi,200)
y1 = np.cos(x)
y2 = np.sin(x)
y3 = np.tanh(x) 
y4 = np.exp(x)

plt.figure(figsize=(18,8))

plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)

plt.show()
```

```python
#Matplotlib gridspec     參考:https://wizardforcel.gitbooks.io/matplotlib-user-guide/content/3.3.html
import matplotlib.pyplot as plt
import matplotlib.gridspec as grd
import numpy as np

x = np.linspace(-np.pi,np.pi,200)
y1 = np.cos(x)
y2 = np.sin(x)
y3 = np.tanh(x) 
y4 = np.exp(x)

plt.figure(figsize = (18,4))

G = grd.GridSpec(2,3)

axes1 = plt.subplot(G[0,:])
plt.plot(x,y1)

axes2 = plt.subplot(G[1,0])
plt.plot(x,y2)

axes3 = plt.subplot(G[1,1])
plt.plot(x,y3)

axes4 = plt.subplot(G[1,2])
plt.plot(x,y4)

plt.show()
```

```python
#Matplotlib pyplot.axes
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi,np.pi,200)
y1 = np.cos(x)
y2 = np.sin(x)

plt.figure(figsize = (18,4)) #畫布的大小為長18cm高4cm

#依照[left,bottom,width,height] 設定百分比 
plt.axes([.1,.1,.8,.8])
plt.plot(x,y1)

plt.axes([.4,.2,.2,.2]) #left,bottom 離大圖左及下的比例 width,height 小圖的寬高比例 
plt.plot(x,y2)


plt.show()
```

```python
#Matplotlob Scatter Chart
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,1,20) #0-1之間20個
y = np.sqrt(x) + 0.3*np.random.rand(20)
p = np.poly1d([1,.2,.1]) #numpy.poly1d 產生多項式 x^2+0.2x+0.1
f = np.poly1d(np.polyfit(x,y,2)) #numpy.polyfit 產生擬合的二階多項式係數 
f1 = np.poly1d(np.polyfit(x,y,3)) #numpy.polyfit 產生擬合的三階多項式係數

plt.scatter(x,y,s=20,c=y,cmap='cool',alpha=.6) # s表示點的大小，c是color，marker就是點的形状 alpha=透明度

#plt.plot(x,y,'ro') #畫紅點

plt.plot(x,f(x),color='g',label='f(x)',linestyle='-.') #綠的

plt.plot(x,f1(x),color='m',label='f1(x)',linestyle=':')

plt.plot(x,p(x),color='b',label='p(x)',linestyle='--') #藍的

plt.legend(loc='best')

plt.show()
```

```python
#Matplotlob Bar Chart
import matplotlib.pyplot as plt
import numpy as np

X = np.arange(20)
Y = np.random.uniform(0.5,1.0,20)

for x,y in zip(X,Y):
    #ha:horizontal alignment
    #va:vertical alignment
    plt.text(x,y,'%.2f'%y,ha='center',va='bottom',fontsize=6) #水平為置中對齊,垂直為靠底部對齊

plt.bar(X,Y,facecolor='#9999ff',edgecolor='white') #facecolor設置主體顏色，edgecolor設置邊框顏色為白色

plt.show()
```

```python
#Matplotlob Pie Chart
import matplotlib.pyplot as plt
import numpy as np

labels = ['Q1','Q2','Q3','Q4']
sizes = [20,30,40,10]
explode = (0,0.1,0,0)

plt.axis('equal') #圓形

#cmap: 顏色圖譜（colormap)  顏色:https://www.cnblogs.com/denny402/p/5122594.html
cmap = plt.cm.spring  #magma黑-紅-白
colors = cmap(np.linspace(0., 1., len(labels)*2))

wedges, texts, autotexts = plt.pie(sizes, colors=colors, explode=explode, labels=labels,
                                   autopct='%1.2f%%', shadow=False, startangle=90) #autopct用來顯示百分比 startangle默認x第一塊餅和水平面的角度不固定。自定義第一塊餅圖和水平面的角度。

plt.setp(autotexts,size=8,weight='bold',color='white')
plt.setp(texts,size=8,weight='bold')
plt.title('Pie Chart', weight='bold') #bold為粗體

plt.show()
```

```python
#Matplotlob Grid與Date
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import pandas as pd

x_raw = pd.period_range(pd.datetime.now(),periods=30,freq='d')
# print(x_raw)
x = x_raw.to_timestamp().to_pydatetime()
# print(x)

#plt.figure(figsize=(18,10)) 調整張圖大小

y1 = np.random.randint(50,80,30)
y2 = np.random.randint(20,45,30)

g = plt.gca() #取得目前座標軸

for tick in g.xaxis.get_majorticklabels():   
    tick.set_fontsize(6)

for tick in g.yaxis.get_majorticklabels():   
    tick.set_fontsize(6)

g.xaxis.set_major_formatter(mdates.DateFormatter('%y-%m-%d'))
plt.gcf().autofmt_xdate() #自動產生日期格式

plt.title('Random Trends')
plt.xticks(fontsize=12) #設定刻度字體大小
plt.yticks(fontsize=12)
plt.grid(True)

plt.plot(x, y1)
plt.plot(x, y2)

plt.show()
```

```python
#Matplotlib 3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np



# set up a figure twice as wide as it is tall
fig = plt.figure(figsize=plt.figaspect(0.5)) #figaspect因為是3D圖 會要一個傾斜的角度

# 第一個圈子
ax = fig.add_subplot(1, 2, 1, projection='3d')

X = np.arange(-5, 5, 0.25) #-5~5間隔為0.25
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y) #產生網格的座標
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)  #rstride cstride 3D圖上格子的長寬度  # coolwarm冷到熱的色調 #antialiased鋸齒

ax.set_zlim(-1.01, 1.01) #Z軸的範圍

fig.colorbar(surf, shrink=0.5, aspect=10) #旁邊那個像溫度計的東西  surf對應3D圖的顏色


# 第二個圈子
from mpl_toolkits.mplot3d.axes3d import get_test_data

ax = fig.add_subplot(1, 2, 2, projection='3d')

X, Y, Z = get_test_data(0.05)

ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10)

plt.show()
```
